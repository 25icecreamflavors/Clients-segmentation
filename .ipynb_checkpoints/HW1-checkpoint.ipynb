```python
df_agg = df.agg(
    [
        "nunique",
        (lambda x: x.nunique() / len(x) * 100),
        (lambda x: x.isna().sum()),
        (lambda x: x.isna().sum() / len(x) * 100),
        (lambda x: x.isin([0]).sum()),
        (lambda x: x.isin([0]).sum() / len(x) * 100),
    ]
).transpose()

df_agg.columns = [
    "Unique",
    "Percentage of unique",
    "NaNs",
    "Percentage of NaNs",
    "Null values",
    "Percentage of nulls",
]
df_agg
```

> ## Часть 1.2 - Делаем проверку на некорректные данные

> Видим, что присутствует некорректное значение "-1" в столбце, который отвечает за время, прошедшее после последнего выданного кредита, при этом у таких клиентов время, прошедшее после первого выданного кредита, положительное. Эти данные некорректные и их немного, мы не будем их использовать для обучения и тестирования моделей.

```python
df[df.min_MnthAfterLoan == -1]
```

> Всё верно, видим, что нет данных, где после первого выданного кредита прошло меньше времени, чем после последнего выданного:

```python
len(df[df.max_MnthAfterLoan < df.min_MnthAfterLoan])
```

> Здесь видим, что в образовании, а также в размере зарплаты у некоторых стоит "no answer", а не `NaN`, заменим эти значения на `NaN`, который обработаем позже вместе с другими пропусками.

```python
len(df[df.EDUCATION == "*n.a.*"])
```

```python
df.loc[df.EDUCATION == "*n.a.*", "EDUCATION"] = np.nan
```

```python
df.loc[df.EMPL_SIZE == "*n.a.*", "EMPL_SIZE"] = np.nan
```

* По другим столбцам данные достаточно корректные. Где-то присутствует неравномерное распределение и некие выбросы, но сейчас они не кажутся критичными. Попробуем вначале учесть те данные во время обучения, так как они тоже могут оказаться полезными для формирования сегментов. При плохих результатах можно будет использовать шкалу логарифма, чтобы уменьшить значение выбросов.

* Также в данных около половин строк имеют `NaN`, мы не можем удалить столько данных, поэтому заменить пропуски на определённое значение, которое не будет ухудшать качество обучения. 

> ## Часть 1.3 - Построение витрины для сегментации

> Заполним пропуски глобальной отрицательной константой. Так как значения признаков положительные, то такое значение не смешается с другими и не ухудшит результат. Для категориальных признаков сделаем так же, потому что onehot кодирование такие значения просто выделит в одну категорию, что нам и нужно. Удалим некорректные строки, которые описали в прошлой части.

```python
df_model = df[df.min_MnthAfterLoan != -1]
```

```python
df_model = df_model.fillna(-68)
```

> Некоторые признаки из данных числовых, на самом деле, являются категориальаными, так как обозначают категорию, а не определённую меру. Запишем их в список категориальных и обработаем с помощью onehot кодирования в будущем.

Список таких признаков: 'if_zalog', 'BANKACCOUNT_FLAG', 'dlq_exist', 'thirty_in_a_year', 'sixty_in_a_year', 'ninety_in_a_year', 'thirty_vintage', 'sixty_vintage', 'ninety_vintage'.

```python
categorical = df.select_dtypes(["object"]).columns.values
numeric = df.select_dtypes(["float64"]).columns.values
print("Числовые признаки:", numeric)
print("\nКатегориальные признаки:", categorical)
```

```python
categorical = np.append(
    categorical,
    [
        "if_zalog",
        "BANKACCOUNT_FLAG",
        "dlq_exist",
        "thirty_in_a_year",
        "sixty_in_a_year",
        "ninety_in_a_year",
        "thirty_vintage",
        "sixty_vintage",
        "ninety_vintage",
    ],
)
print(categorical)
```

```python
numeric = numeric[
    ~np.isin(
        numeric,
        [
            "INCOME_BASE_TYPE",
            "CREDIT_PURPOSE",
            "SEX",
            "EDUCATION",
            "EMPL_TYPE",
            "EMPL_SIZE",
            "EMPL_PROPERTY",
            "EMPL_FORM",
            "FAMILY_STATUS",
            "if_zalog",
            "BANKACCOUNT_FLAG",
            "dlq_exist",
            "thirty_in_a_year",
            "sixty_in_a_year",
            "ninety_in_a_year",
            "thirty_vintage",
            "sixty_vintage",
            "ninety_vintage",
        ],
    )
]
numeric = np.append(numeric, ["INSURANCE_FLAG"])
print(numeric)
```
